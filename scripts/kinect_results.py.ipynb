```python
def get_poses(cup_name):

    cup_indices = np.array([cup_name in s for s in arr[:,0]])

    # print('cup names', arr[:, 0][cup_indices])
    real_poses = np.array(arr[:, 1:3][cup_indices], dtype=np.float32)
    pred_poses = np.array(arr[:, 3:5][cup_indices], dtype=np.float32)

    #distances = np.linalg.norm(real_poses - pred_poses, axis=0)
    
    return real_poses, pred_poses
```

```python
cup_names = ['rocket', 'karol', 'gray', 'can', 'blue', 'subway', 'yellow', 'mirror', 'red']
len(cup_names)
```

```python
fig, axes = plt.subplots(3, 3, sharex=True, figsize=[30, 30])

for i, cup_name in enumerate(cup_names):
    ax = axes[int(i/3)][i%3]
    
    real_poses, pred_poses = get_poses(cup_name)
    ax.scatter(real_poses[:, 0], real_poses[:, 1], c='r', label='real')
    ax.scatter(pred_poses[:, 0], pred_poses[:, 1], c='b', label='pred')
    ax.set_title(cup_name)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()
```
